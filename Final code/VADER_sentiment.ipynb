## Calculate the correlation coefficient

### Prepare the data

### Do the calculation for correlation

```python
def calculate_corelation(sentiment_scores, stock_prices):

    # Calculate the returns
    stock_returns = [(stock_prices[i] - stock_prices[i-1]) / stock_prices[i-1] for i in range(1, len(stock_prices))]
    # Prepend 0 at the beginning of the array
    stock_returns = np.insert(stock_returns, 0, 0)

    # Calculate the covariance matrix
    corelation_matrix = np.corrcoef(sentiment_scores, stock_returns)

    return corelation_matrix

# Example data (replace with your actual data)
sentiment_scores = merged_df["sentiment_score"].tolist()
stock_prices = merged_df["Adj Close"].tolist()

# Calculate the corelation matrix
corelation_matrix = calculate_corelation(sentiment_scores, stock_prices)

print("corelation Matrix:")
print(corelation_matrix)
```
